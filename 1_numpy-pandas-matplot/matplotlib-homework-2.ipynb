```python
data = sns.load_dataset("iris")
data.head()
# 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
```

```python
import numpy as np
import pandas as pd
```

```python
plt.scatter(data.sepal_length, data.sepal_width, marker='^')
plt.scatter(data.petal_length, data.petal_width, marker='o')
```

```python
help(data.groupby('species'))
```

```python
grouped = data.groupby('species')
subplot_index = 221
for key, group in grouped :
    #print(key)
    #print(group)
    plt.subplot(subplot_index)
    plt.title(key)
    plt.scatter(group.sepal_length, group.sepal_width, marker='^')
    plt.scatter(group.petal_length, group.petal_width, marker='o')
    subplot_index = subplot_index + 1
```

```python
mean_species = data.groupby('species').mean()
```

```python
mean_species
```

```python
plt.subplot(221)
plt.title('sepal_length')
plt.bar(mean_species.index, mean_species.sepal_length, color='r')

plt.subplot(222)
plt.title('sepal_width')
plt.bar(mean_species.index, mean_species.sepal_width, color='g')

plt.subplot(223)
plt.title('petal_length')
plt.bar(mean_species.index, mean_species.petal_length, color='b')

plt.subplot(224)
plt.title('petal_width')
plt.bar(mean_species.index, mean_species.petal_width, color='c')
```
